```python
## Image flipping
import cv2
import numpy as np
import os
from glob import glob

with open('images.txt') as f:
    images = f.readlines()
with open('labels.txt') as f:
    labels = f.readlines()

images = [x[:-1] for x in images]
labels = [x[:-1] for x in labels]

for i in range(len(images)):
    print(i+1,"/" ,str(len(images)))
    img_rbg = cv2.imread(images[i])
    img_rbg = cv2.flip(img_rbg, 1)
    img_label = cv2.imread(labels[i])
    img_label = cv2.flip(img_label, 1)
    img_rbg_path = images[i].rsplit('data2',1)[0] + "mirrored" + images[i].rsplit('data2',1)[1]
    img_label_path = labels[i].rsplit('data2',1)[0] + "mirrored" + labels[i].rsplit('data2',1)[1]

    cv2.imwrite(img_rbg_path,img_rbg)
    cv2.imwrite(img_label_path,img_label)
```

```python
# Gaussian blur images
import cv2
import numpy as np
import os
from glob import glob

with open('images.txt') as f:
    images = f.readlines()
with open('labels.txt') as f:
    labels = f.readlines()

images = [x[:-1] for x in images]
labels = [x[:-1] for x in labels]

for i in range(len(images)):
    print(i+1,"/" ,str(len(images)))
    img_rbg = cv2.imread(images[i])
    img_rbg = cv2.GaussianBlur(img_rbg,(7,7),0)
    img_label = cv2.imread(labels[i])
    img_rbg_path = images[i].rsplit('data2',1)[0] + "gaussianblur" + images[i].rsplit('data2',1)[1]
    img_label_path = labels[i].rsplit('data2',1)[0] + "gaussianblur" + labels[i].rsplit('data2',1)[1]

    cv2.imwrite(img_rbg_path,img_rbg)
    cv2.imwrite(img_label_path,img_label)
```

```python
# brightness
import cv2
import numpy as np
import os
from glob import glob
from PIL import Image, ImageEnhance

with open('images.txt') as f:
    images = f.readlines()
with open('labels.txt') as f:
    labels = f.readlines()

images = [x[:-1] for x in images]
labels = [x[:-1] for x in labels]

for i in range(len(images)):
    print(i+1,"/" ,str(len(images)))
    img_rbg = Image.open(images[i])
    img_label = cv2.imread(labels[i])
    img_rbg = ImageEnhance.Brightness(img_rbg).enhance(1.5)

    img_rbg_path = images[i].rsplit('data2',1)[0] + "brightness+50" + images[i].rsplit('data2',1)[1]
    img_label_path = labels[i].rsplit('data2',1)[0] + "brightness+50" + labels[i].rsplit('data2',1)[1]

    img_rbg.save(img_rbg_path)
    cv2.imwrite(img_label_path,img_label)
```

```python
# glare
import cv2
import numpy as np
import os
from glob import glob

with open('images.txt') as f:
    images = f.readlines()
with open('labels.txt') as f:
    labels = f.readlines()

images = [x[:-1] for x in images]
labels = [x[:-1] for x in labels]

for i in range(len(images)):
    print(i+1,"/" ,str(len(images)))
    img_rbg = cv2.imread(images[i])
    gray = cv2.cvtColor(img_rbg, cv2.COLOR_BGR2GRAY)
    mask = cv2.threshold(gray, 180, 220, cv2.THRESH_BINARY)[1]
    img_rbg = cv2.inpaint(img_rbg, mask, 21, cv2.INPAINT_TELEA) 




    img_label = cv2.imread(labels[i])
    img_rbg_path = images[i].rsplit('data2',1)[0] + "glare" + images[i].rsplit('data2',1)[1]
    img_label_path = labels[i].rsplit('data2',1)[0] + "glare" + labels[i].rsplit('data2',1)[1]

    cv2.imwrite(img_rbg_path,img_rbg)
    cv2.imwrite(img_label_path,img_label)




```

```python
# glare
import cv2
import numpy as np
import os
from glob import glob

with open('images.txt') as f:
    images = f.readlines()
with open('labels.txt') as f:
    labels = f.readlines()

images = [x[:-1] for x in images]
labels = [x[:-1] for x in labels]
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])


for i in range(len(images)):
    print(i+1,"/" ,str(len(images)))
    img_rbg = cv2.imread(images[i])
    img_rbg = cv2.filter2D(src=img_rbg, ddepth=-1, kernel=kernel)
    img_label = cv2.imread(labels[i])
    img_rbg_path = images[i].rsplit('data2',1)[0] + "sharpen" + images[i].rsplit('data2',1)[1]
    img_label_path = labels[i].rsplit('data2',1)[0] + "sharpen" + labels[i].rsplit('data2',1)[1]

    cv2.imwrite(img_rbg_path,img_rbg)
    cv2.imwrite(img_label_path,img_label)
```
