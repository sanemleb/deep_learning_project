```python
import numpy as np
import matplotlib.pyplot as plt
import random
 
  
# creating the dataset
data = {'Background', 'Front Door', 'Back Door','Front Fender','Body','Front Bumper','Hood','Rear Bumper','Trunk','Combined'}

import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
model_1 = [96.93, 92.69, 92.95, 95.7, 88.69, 86.65, 94.49, 92.89, 94.49, 92.86]
model_2 = [97.48, 93.24, 91.41, 94.33, 87.57, 85.62, 94.65, 94.89, 96.82, 92.89]
model_3 = [97.52, 94.13, 91.71, 96.5, 87.47, 85.13, 95.83, 94.02, 98.63, 93.55]
# Set position of bar on X axis
br1 = np.arange(len(model_1))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
# Make the plot
plt.rc('axes', axisbelow=True)
plt.bar(br1, model_1, color ='#FE4A49', width = barWidth,
        edgecolor ='grey', label ='Original dataset',zorder=1)
plt.bar(br2, model_2, color ='#04052E', width = barWidth,
        edgecolor ='grey', label ='Generally optimized dataset',zorder=1)
plt.bar(br3, model_3, color ='#65AFFF', width = barWidth,
        edgecolor ='grey', label ='Weak point optimized dataset',zorder=1)
plt.grid(zorder=0,linestyle='--')

 
# Adding Xticks
plt.xlabel('Feature', fontweight ='bold', fontsize = 15)
plt.ylabel('Dice score [%]', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(model_1))],['Background', 'Front Door', 'Back Door','Front Fender','Body','Front Bumper','Hood','Rear Bumper','Trunk','Combined'])
plt.ylim(82,100)
# plt.title("Dice score overview for each class",fontweight ='bold', fontsize = 20)
plt.legend()
# plt.show()
plt.savefig('Bar_plot.png')
```

```python
# Load
model_1 = np.load('Gabriel_dataset_1_0_history_test.npy',allow_pickle='TRUE').item()
model_2 = np.load('Gabriel_dataset_3_0_history_test.npy',allow_pickle='TRUE').item()
model_3 = np.load('kristian_history_test.npy',allow_pickle='TRUE').item()

plt.figure(1)
plt.plot(model_1["loss"],color ='#FE4A49',label= "Original dataset",marker = "o",markersize=3)
plt.plot(model_2["loss"],color ='#04052E',label= "Generally optimized dataset",marker = "o",markersize=3)
plt.plot(model_3["loss"],label= "Weakpoint optimized dataset",marker = "o",markersize=3)
plt.grid()
plt.legend(loc="upper right")
plt.title("Training Loss")
plt.ylabel("Training Loss")
plt.xlabel("Epoch")
# plt.show()
plt.savefig('Training_Loss.png')

plt.figure(2)
plt.plot(model_1["accuracy"],color ='#FE4A49',label= "Original dataset",marker = "o",markersize=3)
plt.plot(model_2["accuracy"],color ='#04052E',label= "Generally optimized dataset",marker = "o",markersize=3)
plt.plot(model_3["accuracy"],color ='#65AFFF',label= "Weakpoint optimized dataset",marker = "o",markersize=3)
plt.legend(loc="lower right")
plt.grid()
plt.title("Training Accuracy")
plt.ylabel("Training Accuracy")
plt.xlabel("Epoch")
# plt.show()
plt.savefig('Training_Accuracy.png')

plt.figure(3)
plt.plot(model_1["val_loss"],color ='#FE4A49',label= "Original dataset",marker = "o",markersize=3)
plt.plot(model_2["val_loss"],color ='#04052E',label= "Generally optimized dataset",marker = "o",markersize=3)
plt.plot(model_3["val_loss"],color ='#65AFFF',label= "Weakpoint optimized dataset",marker = "o",markersize=3)
plt.title("Validation Loss")
plt.legend(loc="upper right")
plt.grid()
plt.ylabel("Validation Loss")
plt.xlabel("Epoch")
# plt.show()
plt.savefig('Validation_Loss.png')


plt.figure(4)
plt.plot(model_1["val_accuracy"],color ='#FE4A49',label= "Original dataset",marker = "o",markersize=3)
plt.plot(model_2["val_accuracy"],color ='#04052E',label= "Generally optimized dataset",marker = "o",markersize=3)
plt.plot(model_3["val_accuracy"],color ='#65AFFF',label= "Weakpoint optimized dataset",marker = "o",markersize=3)
plt.title("Validation Accuracy")
plt.legend(loc="lower right")
plt.grid()
plt.ylabel("Valication Accuracy")
plt.xlabel("Epoch")
# plt.show()
plt.savefig('Validation_Accuracy.png')
```
